```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import networkx as nx
import copy
```

```python
n = 5
shape = (10,) * n
tensor = torch.randn(shape)
```

```python
def merge_nodes(G, node1, node2, new_node, tensor):
    """
    Merge node1 and node2 into new_node in graph G.
    
    Parameters:
        G: networkx.Graph
        node1, node2: nodes to merge
        new_node: the resulting merged node
        combine_attrs_func: function to combine node attributes (optional)
    """
    
    new_attrs = {"tensor": tensor}  # Simple override

    # Add the new node with combined attributes
    G.add_node(new_node, **new_attrs)
    
    # Connect all neighbors of node1 and node2 to new_node
    for n in list(G.neighbors(node1)) + list(G.neighbors(node2)):
        if n != node1 and n != node2:
            G.add_edge(new_node, n)
    
    # Remove original nodes
    G.remove_nodes_from([node1, node2])
```

```python
class TN():
    def initTensor(self, shape):
        return torch.randn(shape)
    
    def __init__(self, G, sizes, ranks):
        if len(G.edges) != len(ranks):
            raise Exception("Rank length must be equal to the number of edges")
        if len(G.nodes) != len(sizes):
            raise Exception("Sizes length must be equal to the number of nodes")
        self.data = nx.MultiGraph()
        i = 0
        for node in G.nodes:
            shape = [sizes[i]]
            self.data.add_node(node, shape=shape)
            i += 1
        i = 0
        for edge in G.edges:
            self.data.nodes[edge[0]]['shape'].append(ranks[i])
            self.data.nodes[edge[1]]['shape'].append(ranks[i])
            self.data.add_edge(edge[0], edge[1], rank=ranks[i], from_index=len(self.data.nodes[edge[0]]['shape']) - 1, to_index=len(self.data.nodes[edge[1]]['shape']) - 1)
            i += 1
        i = 0
        for node in G.nodes:
            self.data.nodes[node]['tensor'] = self.initTensor(self.data.nodes[node]['shape'])
        self.m = len(self.data.nodes) + 1
        print(self.data.edges(data=True))

    def merge_nodes(self, node1, node2):
        t1 = self.data.nodes[node1]['tensor']
        t2 = self.data.nodes[node2]['tensor']
        froms = []
        tos = []
        for key, val in self.data.get_edge_data(node1, node2).items():
            froms.append(val['from_index'])
            tos.append(val['to_index'])
        for key, val in self.data.get_edge_data(node2, node1).items():
            tos.append(val['from_index'])
            froms.append(val['to_index'])
        
        t = torch.tensordot(t1, t2, dims=([froms, tos]))

        # Create new node
        self.data.add_node(self.m, tensor = t, shape = t.shape)

        for e in self.data.edges():
            if e[0] in [node1, node2]:
                e[0] = self.m
                # Actualitzar el from_index
                c = 0
                for i in range(len(froms)):
                    if froms[i] < e[2]['from_index']
                    c += 1
                e[2]['from_index'] -= c
            if e[1] in [node1, node2]:
                e[1] = self.m
                # Actualitxar el to_index
                c = 0
                for i in range(len(froms)):
                    if tos[i] < e[2]['from_index']
                    c += 1
                e[2]['to_index'] += len(t1.shape)-c
        
        self.data.remove_nodes_from([node1, node2])
        self.m += 1

    def edges(self):
        return self.data.edges
    
    def eval(self):
        J = copy.deepcopy(self)
        while len(J.edges()) > 1:
            edge = list(J.edges())[0]
            J.merge_nodes(edge[0], edge[1])
        
```

```python
# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1)])

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()

tn = TN(G, [10, 10, 10], [5,5,5])
tn.eval()
```
