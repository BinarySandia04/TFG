```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x * np.sin(200 * (x + 1)**2)
def g(x):
    return np.sin(250 * x ** 3) * np.cos(150 * x)
def h(x):
    return np.sin(1000 * x ** 3)
```

```python
x = np.linspace(0, 1, 1000)
y = f(x)

plt.plot(x, y, label="f(x) = x sin(200 (x+1)^2)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig('function_graphs/f.png')
plt.show()

plt.clf()
x = np.linspace(0, 1, 1000)
y = g(x)

plt.plot(x, y, label="g(x) = 250x^3 * cos(150x)", color="red")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.savefig('function_graphs/g.png')
plt.show()

plt.clf()
x = np.linspace(0, 1, 1000)
y = h(x)

plt.plot(x, y, label="h(x) = sin(1000x^3)", color="green")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.savefig('function_graphs/h.png')
plt.show()
```
