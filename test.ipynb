```python
# Load image using PIL

# image = Image.open('doradito.png')  # Replace with your image path
# image_shape = (4, 2, 131, 4, 3, 41)

image = Image.open('fruits-llr.png').convert("L")

data = [
    {
        "shape": (8, 8, 4, 8, 8),
        "edges": [(1, 2), (2, 3), (3, 4), (4, 5)],
        "ranks": [8, 8, 8, 8],
        "iters": 200,
        "params": [0.25, 0.5, 1, 2, 4, 8, 12],
        "max_rank": 40
    },
    {
        "shape": (8, 8, 4, 8, 8),
        "edges": [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)],
        "ranks": [8, 8, 8, 8, 8],
        "iters": 200,
        "params": [0.25, 0.5, 1, 2, 4, 8, 12],
        "max_rank": 25
    },
    {
        "shape": (8, 8, 4, 8, 8),
        "edges": [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (3, 5), (1, 4)],
        "ranks": [6, 6, 6, 6, 6, 6, 6],
        "iters": 200,
        "params": [0.25, 0.5, 1, 2, 4, 8, 12],
        "max_rank": 18
    },
    {
        "shape": (8, 8, 4, 8, 8),
        "edges": [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 4), (5, 2), (3, 5)],
        "ranks": [4,4,4,4, 4,4,4,4],
        "iters": 200,
        "params": [0.25, 0.5, 1, 2, 4, 8, 12],
        "max_rank": 12
    },
    {
        "shape": (8, 8, 4, 8, 8),
        "edges": [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (5, 2), (5, 3), (4, 1), (4, 2), (3, 1)],
        "ranks": [3,3,4,3,3,4,3,3,4,3],
        "iters": 200,
        "params": [0.25, 0.5, 1, 2, 4, 8, 12],
        "max_rank": 10
    }
]

i = 1
for d in data:
    print(d)
    image_shape = d["shape"]
    
    # Define transformation: convert image to tensor
    transform = transforms.ToTensor()
    # device = "cpu"
    
    # Apply transformation
    image_tensor = transform(image).squeeze(0)
    
    # Optional: check shape
    # print(image_tensor.shape)  # (C, H, W)
    objective = image_tensor.reshape(image_shape).to(device)
    
    
    G = nx.Graph()
    G.add_edges_from(d["edges"])
    
    nx.draw(G, with_labels=True)
    plt.savefig("results/img-compression/graph-" + str(i) + ".png", format="PNG")
    plt.show()
    plt.clf()
    
    ntn = TN(G, list(objective.shape), d["ranks"])

    print("Without TnALE:")
    print("Original: " + str(torch.numel(objective)))
    print("Tensor Network: " + str(ntn.get_tn_size()))
    print("CR: " + str(torch.numel(objective) / ntn.get_tn_size()))
    
    start_time = time.time()
    
    rel_err, graph = TN.als(ntn, objective, 1e-3, iter_num=d["iters"], print_iters=True)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Time: {elapsed_time:.2f}s" + "\n")
    
    # If image_tensor is (C, H, W), convert to (H, W, C)
    image_np = TN.eval(ntn).reshape(image_tensor.shape).cpu().numpy()
    
    # Show image
    norm = (image_np - np.min(image_np)) / (np.max(image_np) - np.min(image_np))
    plt.imshow(norm)
    plt.imsave("results/img-compression/fruits-comp" + str(i) + ".png", norm)
    plt.clf()

    graph_data = {}
    graph_data["Original"] = {"data": graph[1]}

    print("Rel Error: " + str(rel_err))

    with open("results/img-compression/data" + str(i) + ".txt", "w") as f:
        f.write("Arbitrary structure\n")
        f.write("Shape: " + str(d["shape"]) + "\n");
        f.write("Edges: " + str(d["edges"]) + "\n");
        f.write("Ranks: " + str(d["ranks"]) + "\n\n");
        f.write("Original: " + str(torch.numel(objective)) + "\n")
        f.write("Tensor Network: " + str(ntn.get_tn_size()) + "\n")
        f.write("CR: " + str(torch.numel(objective) / ntn.get_tn_size()) + "\n")
        f.write("Rel Error: " + str(rel_err) + "\n")
        f.write("Iters: " + str(d["iters"]) + "\n")
        f.write(f"Time: {elapsed_time:.2f}s" + "\n")

    # ---------------------------------------------------------------------------------
    del ntn

    for tuning_param in d["params"]:  
        G = nx.Graph()
        G.add_edges_from(d["edges"])
        
        print("Tuning param " + str(tuning_param))
        print("Doing TnALE Initial Search...")
        
        start_time = time.time()
        G, R = TN.tn_ale(G, d["ranks"], 5, 3, objective, print_iters=True, update_graph=True, tuning_param=tuning_param, max_rank=d["max_rank"])
        end_time = time.time()
        tnale_first_time = end_time - start_time
        
        print("Doing TnALE Concrete Search...")
        
        start_time = time.time()
        G, R = TN.tn_ale(G, R, 2, 3, objective, print_iters=True, tuning_param=tuning_param, max_rank=d["max_rank"])
        end_time = time.time()
        tnale_second_time = end_time - start_time
        
        print("Found structure has ranks " + str(R))

        print("Found graph:")
        nx.draw(G, with_labels=True)
        plt.savefig("results/img-compression/graph-" + str(i) + "-tnale-" + str(tuning_param) + ".png", format="PNG")
        plt.show()
        plt.clf()
        
        
        ntn = TN(G, list(objective.shape), R)
        
        print("With TnALE:")
        print("Original: " + str(torch.numel(objective)))
        print("Tensor Network: " + str(ntn.get_tn_size()))
        print("CR: " + str(torch.numel(objective) / ntn.get_tn_size()))
        
        start_time = time.time()
        rel_err, graph = TN.als(ntn, objective, 1e-3, iter_num=d["iters"], print_iters=True)
        end_time = time.time()

        als_time = end_time - start_time
        
        print(f"Time: {elapsed_time:.2f}s" + "\n")
        
        # If image_tensor is (C, H, W), convert to (H, W, C)
        image_np = TN.eval(ntn).reshape(image_tensor.shape).cpu().numpy()
        
        # Show image
        norm = (image_np - np.min(image_np)) / (np.max(image_np) - np.min(image_np))
        plt.imshow(norm)
        plt.imsave("results/img-compression/fruits-comp" + str(i) + "-ale-" + str(tuning_param) + ".png", norm)
        plt.clf()
    
        print("Rel Error: " + str(rel_err))
    
        with open("results/img-compression/data" + str(i) + "-ale-" + str(tuning_param) + ".txt", "w") as f:
            f.write("TnALE found structure:\n")
            f.write("Shape: " + str(d["shape"]) + "\n");
            f.write("Edges: " + str(d["edges"]) + "\n");
            f.write("Ranks: " + str(R) + "\n\n");
            f.write("Original: " + str(torch.numel(objective)) + "\n")
            f.write("Tensor Network: " + str(ntn.get_tn_size()) + "\n")
            f.write("CR: " + str(torch.numel(objective) / ntn.get_tn_size()) + "\n")
            f.write("Rel Error: " + str(rel_err) + "\n")
            f.write("Iters: " + str(d["iters"]) + "\n")
            f.write("ALS Time: " + str(als_time) + "\n")
            f.write("TnALE First Batch: " + str(tnale_first_time) + "\n")
            f.write("TnALE Second Batch: " + str(tnale_second_time) + "\n")
            f.write(f"Time: {elapsed_time:.2f}s" + "\n")
        del ntn

        graph_data[str(tuning_param)] = {"data": graph[1]}

    # Print graph
    plt.ylim(0, 0.3)
    plt.xlim(1, 10)
    plt.xticks(range(1, 11))
    for k in graph_data.keys():
        print(k)
        plt.plot(graph_data[k]["data"], label=k)
    plt.legend()
    plt.savefig('results/img-compression/rel_err_fig-' + str(i) + ".png")
    plt.show()

    i += 1
```
